```python
import os
import shutil
import random

BASE_DIR = "datasets/demo_dataset"
TEMP_DIR = "datasets/temp_collection"

def clean_and_split_data(base_dir, temp_dir, train_ratio=0.8, val_ratio=0.1, test_ratio=0.1):
    """splitting in raion 80/10/10 """

    random.seed(42)

    print("some info about process")

    class_names = set()
    all_files = {}


    for root in ['train', 'val']:
        root_path = os.path.join(base_dir, root)
        if os.path.exists(root_path):
            for cls in os.listdir(root_path):
                cls_path = os.path.join(root_path, cls)
                if os.path.isdir(cls_path):

                    class_names.add(cls)

                    os.makedirs(os.path.join(temp_dir, cls), exist_ok=True)

                    if cls not in all_files:
                        all_files[cls] = []

                    for f in os.listdir(cls_path):
                        src_file = os.path.join(cls_path, f)
                        if os.path.isfile(src_file):

                            new_name = f"{root}_{f}"
                            dst_file = os.path.join(temp_dir, cls, new_name)
                            # Faylı müvəqqəti qovluğa köçürürük
                            shutil.copy2(src_file, dst_file)
                            all_files[cls].append(new_name)

    if not class_names:
        print("XƏTA: Sinif qovluqları (moderate, normal, severe) tapılmadı! Skripti dayandırıram.")
        return

    print(f"Toplanan siniflər: {list(class_names)}")

    print("2. Köhnə (təkrar şəkillərlə dolu) qovluqlar silinir və yeniləri yaradılır...")

    # 2. Köhnə qovluqları silirik (test yoxdusa da fərq etməz, yenisini yaradacağıq)
    for folder in ['train', 'val', 'test']:
        full_path = os.path.join(base_dir, folder)
        if os.path.exists(full_path):
            shutil.rmtree(full_path)
        os.makedirs(full_path)

    print("3. Hər bir sinif üzrə şəkillər təsadüfi olaraq 80/10/10 nisbətində bölünür...")

    # 3. Yeni təsadüfi bölüşdürmə
    for cls in class_names:
        files = all_files.get(cls, [])
        if not files:
            continue

        random.shuffle(files)
        n_files = len(files)
        n_train = int(train_ratio * n_files)
        n_val = int(val_ratio * n_files)

        train_files = files[:n_train]
        val_files = files[n_train:n_train + n_val]
        test_files = files[n_train + n_val:]

        # Hədəf qovluqlarını siniflər üçün yaradırıq
        os.makedirs(os.path.join(base_dir, 'train', cls), exist_ok=True)
        os.makedirs(os.path.join(base_dir, 'val', cls), exist_ok=True)
        os.makedirs(os.path.join(base_dir, 'test', cls), exist_ok=True)

        # Faylları müvəqqəti qovluqdan yeni qovluqlara köçürürük
        for target_folder, file_list in [('train', train_files), ('val', val_files), ('test', test_files)]:
            for f in file_list:
                src = os.path.join(temp_dir, cls, f)
                dst = os.path.join(base_dir, target_folder, cls, f)
                shutil.move(src, dst)

    print("4. Müvəqqəti qovluq silinir...")
    shutil.rmtree(temp_dir)

    print(f"\n✅ DATA UĞURLA TƏMİZLƏNDİ və YENİDƏN BÖLÜNDÜ ({int(train_ratio*100)}/{int(val_ratio*100)}/{int(test_ratio*100)}).")
    print("İndi 'train', 'val', və 'test' qovluqlarında REAL, fərqli şəkillər var.")

# Skripti işə salırıq:
clean_and_split_data(BASE_DIR, TEMP_DIR)
```
